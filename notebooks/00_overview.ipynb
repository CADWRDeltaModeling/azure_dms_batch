# Overview 

These examples assume that the batch account and associated storage account have been setup. This can be done by [following the instructions here](https://docs.microsoft.com/en-us/azure/batch/quick-run-python). The information goes into a config file that can be generated by using the generate_blank_config()

This is the index notebook to link to the samples and give an overview of the azure batch samples

* Sample (windows) with a simple executable [sample_submit_dsm2_hydro_version.ipynb](sample_submit_dsm2_hydro_version.ipynb)
* Sample (linux) with a simple executable [sample_submit_dsm2_linux_hydro_version.ipynb](sample_submit_dsm2_linux_hydro_version.ipynb)

The next two samples take longer to run (~ 55 minutes) and were meant more to reflect a real world. They can also be used to benchmark with a real world application

* [sample_submit_dsm2_linux_historical.ipynb](sample_submit_dsm2_linux_historical.ipynb)
* [sample_submit_dsm2_historical.ipynb](sample_submit_dsm2_historical.ipynb)

Another example with a real world example. 
* [sample_submit_C2VSimFG.ipynb](sample_submit_C2VSimFG.ipynb)

An example with master/slave program with networking between the nodes. Also the master task needs to start before slaves can be started by reading the standard output from the master task and using that to start the slave tasks on other nodes

* [sample_submit_beopest.ipynb](sample_submit_beopest.ipynb)

An example with a program with fine-grained MPI and requiring an fast high bandwidth network (Infiniband)
* [sample_submit_hello_schism.ipynb](sample_submit_hello_schism.ipynb)
